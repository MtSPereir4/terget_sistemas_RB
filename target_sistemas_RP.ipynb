##Problema 01: Sequência de Fibonacci

```python
# Entrada do usuário
numero = int(input("Informe um número para verificar se ele pertence à sequência de Fibonacci: "))

# Criando a sequência de Fibonacci
fibonacci_seq = [0 , 1]
# Primeiro termo
n1 = fibonacci_seq[0]
# Segundo termo
n2 = fibonacci_seq[1]
while True:
  # Novo termo
  n = n1 + n2
  # Adicionando o novo termo a sequência
  fibonacci_seq.append(n)
  # Substituindo os dois últimos termos antecessores ao novo termo
  n1 = n2
  n2 = n

  # Condição de parada da seqência de Fibonacci
  if n >= numero:
    break

# Mostrando a sequência de Fibonacci até o número mais próximo do número de entrada
print(f"Sequência de Fibonacci: {fibonacci_seq}")

# Verificando se o número pertence à sequência de Fibonacci
if numero in fibonacci_seq:
  print(f"O número {numero} pertence a sequência de Fibonacci.")
else:
  print(f"O número {numero} não pertence a sequencia de Fibonacci.")
```

##Problema 02: Verificação de um caractere em uma string

```python
# String a ser analisada
string = input("Informe a string desejada: ")

# Verificando a presença do caractere 'a' na string
if 'a' in string.casefold():
  print("O caractere 'a' está na string.")
else:
  print("O caractere 'a' não está na string.")

cont_a = 0
# Percorrendo a string
for i in string.casefold():
  # Contando o número de repetições do caractere 'a'
  if 'a' == i:
    cont_a += 1

print(f"O caractere 'a' aparece {cont_a} vezes na string.")
```

##Problema 03: soma

```python
# Declarando variáveis
indice = 12
k = 1
soma = 0

# Criando o laço de repetição
while k < indice:
  k += 1
  soma += k

# Mostrando o resultado
print(f"A soma vale: {soma}")
```
